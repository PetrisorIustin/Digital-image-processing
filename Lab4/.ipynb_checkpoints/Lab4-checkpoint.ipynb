```python
import cv2
from matplotlib import pyplot as plt
import glob
import numpy as np
from math import floor
from skimage.feature import greycomatrix, greycoprops
from skimage import data

from sklearn.naive_bayes import GaussianNB
%matplotlib notebook
```

```python
def imread(path):
    return cv2.imread(path)[:,:,::-1]
```

```python
images = []
labels = []
features = []
i = 3
for img_path in glob.glob('UPOL/normalized/*.bmp'):
    labels.append(int(i/3))
    img = imread(img_path)
    images.append(img)
    i = i + 1
    
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    glcm = greycomatrix(gray_img, [1], [0, np.pi / 2], 256, symmetric=True, normed=True)
    feature = [greycoprops(glcm, 'correlation')[0, 0], \
               greycoprops(glcm, 'dissimilarity')[0, 0], \
               greycoprops(glcm, 'contrast')[0, 0], \
               greycoprops(glcm, 'homogeneity')[0, 0], \
               greycoprops(glcm, 'energy')[0, 0] \
                  ]
    
    features.append(np.array(feature))
```

```python
def distance(a,b):
    return np.linalg.norm(a-b)
```

```python
clf = GaussianNB()
clf.fit(images, labels)
```

```python
def find_neighbour(idx_image, images, labels):
    index = 0
    min_dist = 10000000
    for i in range(len(images)):
        dist = distance(images[idx_image], images[i])
#         print(dist)
        if min_dist > dist and idx_image != i:
        #if min_dist > dist:
            min_dist = dist
            index = i
            print(index, min_dist)
    return labels[index], index, labels[idx_image]
    
```

```python
find_neighbour(7, images, labels)
```
